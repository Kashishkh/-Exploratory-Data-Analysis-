# Data Preparation and Cleaning

```python
# % of missing values per column
mising_percentage = df.isna().sum().sort_values(ascending=False)/len(df)
mising_percentage
```

```python
mising_percentage.plot(kind='barh')
```

```python
#Handling missing values
df.fillna(df.median(numeric_only=True), inplace=True)
df.isna().sum()
```

```python
# Convert categorical variables to appropriate data types
df['User_ID'] = df['User_ID'].astype(str)
df['Product_ID'] = df['Product_ID'].astype(str)
```

```python
city_sales = df.groupby('City_Category')['Purchase'].sum()

plt.figure(figsize=(8, 8))
plt.pie(city_sales, labels=city_sales.index, autopct='%1.1f%%', startangle=140, textprops={'fontsize': 14})
plt.title('Sales by City Category', fontsize=16)

plt.show()
```

```python
import matplotlib.pyplot as plt

age_groups = df.groupby('Age')['Purchase'].sum()

plt.figure(figsize=(10, 6))
age_groups.plot(kind='bar')
plt.title('Total Sales by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Sales by Occupation
occupation_sales = df.groupby('Occupation')['Purchase'].sum().sort_values(ascending=False)

print("Sales by Occupation (in total):")
for occupation, total_sales in occupation_sales.items():
    print(f"Occupation: {occupation}, Total Sales: {total_sales:,.0f}")

# Plotting
plt.figure(figsize=(12, 8))
occupation_sales.plot(kind='bar')
plt.title('Total Sales by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
gender_sales = df.groupby('Gender')['Purchase'].sum()

plt.figure(figsize=(6, 6))
plt.pie(gender_sales, labels=gender_sales.index, autopct='%1.1f%%', startangle=140, textprops={'fontsize': 14})
plt.title('Sales by Gender', fontsize=16)

plt.show()

age_gender_sales = df.groupby(['Age', 'Gender'])['Purchase'].sum().unstack()

age_gender_sales.plot(kind='bar', figsize=(12, 6))
plt.title('Sales by Age Group and Gender')
plt.xlabel('Age Group')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.show()
```

```python
# Distribution of Purchase
import seaborn as sns # Import the seaborn library and assign it to the alias 'sns'
sns.set_theme()
plt.figure(figsize=(8,6))

sns.histplot(df['Purchase'] ,
             bins=30,
             kde = True )

plt.title("Distribution Of Purchase" ,fontsize = 18)
plt.xlabel('Purchase Amount')
plt.ylabel('Count')
plt.show()
```

```python
# Product Category Distribution
plt.figure(figsize=(10, 5))
sns.countplot(x='Product_Category_1', data=df, order=df['Product_Category_1'].value_counts().index)
plt.title('Product Category Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
# Gender-wise product preferences
plt.figure(figsize=(12, 5))
sns.countplot(x='Product_Category_1', hue='Gender', data=df, order=df['Product_Category_1'].value_counts().index)
plt.title('Gender-wise Product Preferences')
plt.xticks(rotation=45)
plt.show()
```

```python
# Age group vs. product category analysis
plt.figure(figsize=(12, 6))
sns.boxplot(x='Age', y='Product_Category_1', data=df)
plt.title('Age Group vs. Product Category')
plt.xticks(rotation=45)
plt.show()
```

```python
# Marital Status and Purchase behavior
plt.figure(figsize=(7, 5))
sns.boxplot(x='Marital_Status', y='Purchase', data=df)
plt.title('Marital Status vs. Purchase Behavior')
plt.xticks([0, 1], ['Single', 'Married'])
plt.show()
```
